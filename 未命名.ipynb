```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# 生成一组随机数
x = np.linspace(-4, 4, 1000)

# 计算标准正态分布的概率密度函数
y = norm.pdf(x, 0, 1)

# 绘制概率密度函数曲线
plt.plot(x, y)

# 填充曲线左侧和右侧的面积
plt.fill_between(x, 0, y, where=(x < 0), alpha=0.2)
plt.fill_between(x, 0, y, where=(x >= 0), alpha=0.2)

# 添加标题、坐标轴标签等
plt.title('Standard Normal Distribution')
plt.xlabel('X')
plt.ylabel('PDF(X)')

# 显示图像
plt.show()
```
